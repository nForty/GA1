### Import the Airline Passengers data set, preprocess the data, and plot the raw time series.

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil.relativedelta import *
%matplotlib inline

plt.rcParams['figure.figsize'] = [10, 7]
plt.rcParams['font.size'] = 14
plt.style.use('fivethirtyeight')
```

```python
airline = pd.read_csv('../data/airline.csv')
airline.head()
```

```python
airline.tail()
```

```python
airline.columns= ['passengers']
airline['passengers'] = airline.passengers.apply(lambda x: int(x))
```

```python
start = datetime.datetime.strptime("1949-01-01", "%Y-%m-%d")
date_list = [start + relativedelta(months=x) for x in range(0, airline.shape[0])]
airline['date'] = date_list

airline.set_index(['date'], inplace=True)
airline.index.name=None

airline.head() 
```

```python
airline.tail()
```

```python
airline.plot();
```

### Decompose the time series and plot using the `.seasonal_decompose()` function.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(airline, freq = 12)
```

```python
fig = decomposition.plot()
```

### Interpret these plots.

There is a clear upward trend in airline passengers. There is a strong seasonal component.
